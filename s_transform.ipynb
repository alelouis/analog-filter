```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R, C = 500, 470e-6
```

```python
def h(s, R, C):
    return 1/(s * R * C + 1)
```

```python
f = np.logspace(start=-2, stop=2, num=300)
w = 2*np.pi*f
s = 1j*w
h_s = h(s, R, C)
```

```python
plt.figure(dpi=100, figsize = (8, 4))
ax1 = plt.subplot()
plt.semilogx(f, 20*np.log10(np.abs(h_s)), c = 'blue', label = 'Amplitude')
plt.ylabel('Amplitude')
ax1.legend(loc=0)
ax1.minorticks_on()
plt.grid(which="minor", linestyle = ":")
plt.grid(which="major", linestyle = "--")
ax2 = ax1.twinx()
ax2.semilogx(f, np.degrees(np.angle(h_s)), c = 'red', label = 'Phase')
plt.ylabel('Phase')
ax2.legend(loc=3)
plt.xlim([np.min(f), np.max(f)])
```
